```python
with open("daddy.txt", "r", encoding = "utf-8") as f:
    data = f.readlines()

string = data[0]
```

```python
words = string.split(" ")

def next(word: str, words: list) -> list:
    
    lst = []
    
    for i in range(len(words) - 1):
        if words[i] == word:
            lst.append(words[i + 1])
    
    lst = list(set(lst))
    return lst
```

```python
states = {}

for word in words:
    if word not in states:
        lst = next(word, words)
        if lst != []:
            states[word] = lst

print(states)
```

```python
import random

k = "no"
out = ""
for i in range(100):
    k = random.choice(states[k])
    out = out + " " + k

with open("out.txt", "w", encoding = "utf-8") as f:
    f.writelines(out)
```
